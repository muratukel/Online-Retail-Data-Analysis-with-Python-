GIn general, e-commerce datasets are private and therefore publicly available data is difficult to find. However, the [UCI Machine Learning Repository](http://archive.ics.uci.edu/ml/index.php) has maintained this dataset to include real transactions. The dataset can be found under the title "Online Retail".


**Column information:**

**``InvoiceNo:``** Invoice number. A 6-digit integer uniquely assigned to each transaction. If this code starts with the letter 'c', it means cancel.<br>
**``StockCode:``** Product code. a 5-digit integer uniquely assigned to each different product<br>
**``Description:``** Product name<br>
**``Quantity:``** Quantity of each product (item) per transaction. numerical data.<br>
**``InvoiceDate:``** Invoice Date and time. Numeric data, the day and time each transaction was created.<br>
**``UnitPrice:``** Unit price. Numerical data, product price per unit in pounds sterling<br>
**``CustomerID:``** Customer number. A 5-digit integer assigned individually to each customer.<br>
**``Country:``** Country name. The name of the country where each customer resides.<br>

**26.In which countries is the rate of acquiring new customers higher? Which products do new customers usually prefer?**

```python
df["NewCustomer"] = df.groupby('CustomerID')["Date"].transform('min') ==df['Date']

df["NewCustomer"] =df["NewCustomer"].astype(int)
```

```python
new_customer_by_country = df.groupby('Country')["NewCustomer"].sum()

new_customer_by_country.sort_values(ascending=False).reset_index().head(10)
```

```python
# Total number of customers by country

total_customer_by_country =df.groupby('Country')["CustomerID"].nunique()
```

```python
# We calculate acquisition rates.

acquisition_ratio = (
    (new_customer_by_country / total_customer_by_country)
    .fillna(0)
    .reset_index()
)

acquisition_ratio.columns = ['Country', 'AcquisitionRate']

acquisition_ratio = acquisition_ratio.sort_values(by="AcquisitionRate", ascending=False)

print(acquisition_ratio)
```

```python
df.head(3)
```

```python
# Product preference analysis of new customers

new_customer_data = df[df['NewCustomer'] == 1]

new_customer_product_preference = (
    new_customer_data.groupby("Description")["Quantity"]
    .sum()
    .reset_index()
    .sort_values(by="Quantity", ascending=False)
)

new_customer_data

#print(new_customer_product_preference.head(10))
```

**27.How do customer preferences change according to the price ranges of products? Which price ranges are more popular?**

```python
df_filtered.UnitPrice.describe()
```

```python
def segment_price(price):
 
    if price >= 0 and price < 5000:
        return "cheap"
    elif price >= 5000 and price < 10000:
        return "expensive"
    elif price >= 10000 and price < 15000:
        return "very expensive"
    else:
        return "luxury"
    
df_filtered["UnitPriceSegment"] = df_filtered["UnitPrice"].apply(segment_price)
df_filtered["UnitPriceSegment"].value_counts()
```

**28.How do holidays or special occasions (e.g. Christmas, New Year) affect their sales? Which products are more in demand during these periods?**

```python
special_days = ["01/01/2010", "14/02/2010", "08/05/2010", "19/06/2010", "31/10/2010", "25/12/2010",
                 "01/01/2011", "14/02/2011", "08/05/2011", "19/06/2011", "31/10/2011", "25/12/2011"]

special_days_dataset = df[df["Date"].isin(special_days)]

special_days_dataset
```

```python
special_day = special_days_dataset.groupby("Date")["Quantity"].sum().reset_index()

special_day
```

```python
special_dates = [
    ('08/05/2011', 'Mother s Day'),
    ('14/02/2011', 'Valentine s Day'),
    ('19/06/2011', 'Father s Day'),
    ('31/10/2011', 'Halloween')
]

special_day['SpecialDay'] = special_day['Date'].replace(dict(special_dates))
```

```python
special_day
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8, 6))

explode = (0, 0, 0.1, 0) 
plt.pie(
    special_day["Quantity"],
    labels=special_day["SpecialDay"],
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
)
plt.axis('equal')
plt.title('Quantity Distribution On Special Days')
plt.show()
```
